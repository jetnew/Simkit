```python
from data.synthetic import get_normal_data, plot_data

X, y = get_normal_data(1000)
plot_data(X, y)
```

```python
from model.bayesnn import BayesNN

bnn = BayesNN(x_features=2, y_features=1, n_hidden=32)
bnn.fit(X, y, epochs=300)
```

```python
y_hat = bnn.predict(X)
plot_data(X, y_hat)
```

```python
params = {
    "x_features": 2,
    "y_features": 1,
    "epochs": 300,
}
opt_params = {
    'n_hidden': [1,100],
    'n_layers': [1,10],
}

bayesnn = BayesNN.hyperopt(params, opt_params, X, y, trials=30)
```

```python
bayesnn.n_hidden, bayesnn.n_layers
```
